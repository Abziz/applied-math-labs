```python
import numpy as np
import matplotlib.pyplot as plt
```

## 3D plots
Plotting in 3D starts with displaying our plot in 3D:

```python
ax = plt.axes(projection='3d')
plt.show()
```

the `ax` variable allows us to use 3d plotting features. for example setting labels for the `z` axis.

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
plt.show()
```

## Simple 2D function on a 3d plot

Let's see what happens if we plot a 2D function in a 3D plot

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-5,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-5,5,100)
plt.plot(x,x**2)
plt.show()
```

We can see our 2D line graph inside a 3D space. We show `y` as a function of `x`  while fixing `z` to be 0

## Plotting a 3d line

To draw a 3d line, we can define `z` as a function of `x` and `y`.  We can then use `ax.plot3D(x,y,z)` to plot our 3D line.

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-1,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
plt.plot(x,y, x**2 + y**2)
plt.show()
```

Change view angle to see the x axis along the z axis using `ax.view_init`. this is simulating y = 0

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-1,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
ax.view_init(elev=0, azim=90)
plt.plot(x,y, x**2 + y**2)
plt.show()
```

Change view angle to see the y axis along the z axis using `ax.view_init`. This is simulating x = 0

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-1,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
ax.view_init(elev=0, azim=0)
plt.plot(x,y, x**2 + y**2)
plt.show()
```

The way this works, is that `z` acts as a function of `x` and `y` , $z(x,y)=x^2 + y^2$  
for all $x_i \in x$ and $y_i \in y$ it draws the point $z(x_i,y_i)$, hence drawing a single line.

## Drawing all the points

An actual function $z(x,y)$ should plot all combination of $x$ and $y$.  
$z(a,b)$ where $(a,b) \in x \times y$  
to create all combinations of $x$ and $y$ we can use `np.meshgrid`

```python
x = np.array([0,1,2])
y = np.array([0,1,2])
X,Y = np.meshgrid(x,y)
print('X=', X , sep='\n')
print('Y=', Y , sep='\n')
```

Using the values of `X` and `Y` we can plot all combinations of points possible on the grid.

The following code shows the points on the grid where the `z` values are just 0

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-1,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
X,Y = np.meshgrid(x,y)
# using ravel to plot the points.
ax.plot3D(X.ravel(),Y.ravel(),np.zeros(X.shape).ravel(),lineWidth=0,marker='o',markerSize=1)
plt.show()
```

if we plug in an actual value for the Z axis we can see how the points draw a surface. in Pycharm we can play with it and move the view angle to see it more clearly.

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-1,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X,Y = np.meshgrid(x,y)
# using ravel to plot the points.
ax.plot3D(X.ravel(),Y.ravel(),(X**2 + Y**2).ravel(),lineWidth=0,marker='o',markerSize=1)
# we can change lineWidth above to 1 to see whats going on
plt.show()
```

## Drawnig a surface

We can use the meshgrid to draw a surface instead of points and lines using `ax.plot_surface(X,Y,Z)`

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlim((-1,5))
plt.tight_layout() # used to show more of the plot - has nothing to do with 3d
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X,Y = np.meshgrid(x,y)
ax.plot_surface(X,Y,X**2 + Y**2)
plt.show()
```
