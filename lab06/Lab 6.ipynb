## plt.show(block = False) and plt.pause(ms)

When we plot stuff usually our script stop executing and it waits for us to close the plot window.  
We can override this behavior and keep executing code even while plots are showing.

```python
xx = np.linspace(-5, 5, 200)
plt.show(block=False)
thetas = np.linspace(0, 2 * np.pi, 50);
for t in thetas:
    plt.plot(xx, np.cos(xx + t))
    plt.show(block=False)
    plt.pause(0.05)
    
plt.show()
plt.pause(0)
```

using `plt.pause(0)` will hold the plot at the end without closing it and ending the program.

## Clearing figure with plt.clf()

In the previous example we saw how the lines kept stacking on top of each other. We can use `plt.clf()` to clear the figure just before we draw the new one. This example will allways show a single line:

```python
xx = np.linspace(-5, 5, 200)
thetas = np.linspace(0, 2 * np.pi, 100);
for t in thetas:
    plt.clf()
    plt.plot(xx, np.cos(xx + t))
    plt.show(block=False)
    plt.pause(0.05)
plt.pause(0)
```
