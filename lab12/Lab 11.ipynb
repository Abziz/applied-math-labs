```python
df.iloc[0]
```

We can pass a list of integers to retrieve multiple rows

```python
df.iloc[[0,1]]
```

We can also use ptyhon slicing to retrieve multiple rows

```python
df.iloc[3:6]
```

We can use `df.loc` to access rows by index (by default index is a number but it can be changed)

```python
df.loc[0] # same as df.iloc[0] unless index was changed
df.loc[[0,1]] # same as df.iloc[[0,1]] unless index was changed
```

Unlike `df.iloc`, we can use `df.loc` to retrieve specific rows and columns

```python
df.loc[[1,2,3],"Student"]
```

```python
df.loc[5:10,["Gender","LanguageWorkedWith"]]
```

We can also use slicing on column names when working with `df.loc`. notice that the slice is inclusive.

```python
df.loc[5:10,"OpenSourcer":"Country"]
```

## Filtering

We can use conditional operators to retrieve an series of boolean values representing the rows that meet the condition.

```python
df['Country'] == 'Canada'
```

We can see the filtered data by passing the condition into the data frame

```python
df[(df["Country"] == "Canada")]
```

Or we can store the fitler and use it like so:

```python
canada_filter = (df['Country'] == 'Canada')
```

```python
df[canada_filter]
```

We can also use `df.loc` with our filter. We can also get specific columns

```python
df.loc[canada_filter,"YearsCode"]
```

We can use `|` to represent `or` and `&` to represent `and`. Notice the parentheses 

```python
filt = (df["Country"] == "Israel") | (df["Country"] == "Canada")
```

```python
df[filt]
```

```python
filt = (df["Country"] == "Israel") & (df["Age"] <= 20)
```

```python
df.loc[filt,"Age"]
```

We can negate the filter using `~`

```python
df.loc[~filt,"Age"]
```

We can also create filters for acceptable values within a column.  Let's say we want to get all the sallaries from `Israel`,`United States`,`Canada` and `United Kingdom`. Instead of using a long filter we can do the following:

```python
countries = ["Israel","United States","Canada","United Kingdom"]
filt = df['Country'].isin(countries)
```

```python
df.loc[filt]
```

If we look the `LanguageWorkedWith` column, we see a string containing languages seprated by a `;`. We can't simply compare it to `"Python"`. We can use string methods using pandas filters like so:

```python
filt = df["LanguageWorkedWith"].str.contains('Python',na=False) 
```

na = False, ignores rows with missing values. We would get an error if they exist.

```python
df.loc[filt,"LanguageWorkedWith"]
```

## Common data operations

We can use `count()` to count the number of rows. Combined with filters, we can get interesting data

```python
Hobbyist_count = df.loc[(df["Gender"] == "Woman"),"Hobbyist"].count()
Hobbyist_count
```

We can calculate the mean of the columns using `df[column].mean()`

```python
mean_age = df["Age"].mean()
print(f"{mean_age=}")
```

We can calculate the mean of the columns using `df.median()` or `df[column].median()`

```python
median_age = df["Age"].median()
print(f"{median_age=}")
```

We can count values inside a column using `value_count()`

```python
df["Country"].value_counts()
```

We can describe a data frame using `describe()` to return a data frame containing various properties of the data.

```python
df.describe()
```

We can display histograms using `df.hist(...)`. for exmaple:

```python
df.hist(column="ConvertedComp")
```
