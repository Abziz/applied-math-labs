### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

### simple plot

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([1, 2, 4, 8, 16, 32])
plt.plot(x, y)
plt.show()
```

### using linspace

```python
x = np.linspace(-16, 16, 1000)
y = x ** 2
plt.plot(x, y)
plt.show()
```

### sin and  plot margins

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x)
plt.plot(x, y)
plt.margins(x=0, y=0.5)
plt.show()
```

### Multiple graphs in the same plot

```python
x = np.linspace(-10, 10, 1000)
# prefered way
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
# other way
# plt.plot(x, np.sin(x), x, np.cos(x))
plt.margins(x=0, y=0.5)
plt.title("Sin and Cos")
plt.legend()  # takes the inline labels passed as args
plt.show()
```

### Mathematical expressions, labels and legends

```python
x = np.linspace(0, 20, 1000)
np.seterr(divide='ignore', invalid='ignore') # ignore divide by 0 errors
plt.plot(x, np.sqrt(x), label=r"$\sqrt{x}$")
plt.plot(x, np.sin(x) / x, label=r"$\dfrac{sin(x)}{x}$")
plt.margins(x=0, y=0.5)
plt.title("Mathematical expressions (using raw strings)")
plt.legend()
plt.show()
```

### Make graphs pretty with format string

```python
x = np.linspace(-np.pi, np.pi)
# pass a format string in the form: '[marker][line][color]'
# red line
plt.plot(x, np.sin(x), 'r')
# green dashed line
plt.plot(x, np.cos(x), '--g')
# blue x markers without line
plt.plot(x, x, 'xb')

plt.show()
```

### Make graphs pretty with keyword arguments

```python
x = np.linspace(-np.pi, np.pi)
plt.figure(1)
# pass a format string in the form: '[marker][line][color]'
# red line ( defaults to '-')
plt.plot(x, np.sin(x), color='r')
# green dashed line
plt.plot(x, np.cos(x), lineStyle='--', color='g')
# blue circle markers without line
plt.plot(x, x, marker='x', color='b')
plt.show()
```

### Using  sub plots

```python
x = np.linspace(-10, 10, 1000)
plt.figure(1)
plt.subplot(2, 2, 1)
plt.plot(x, np.abs(x), label=r"$|x|$")
plt.grid(True)
plt.legend(loc="upper left")
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label=r"$sin(x)$")
plt.grid(True)
plt.legend(loc="upper left")
plt.subplot(2, 2, 3)
plt.plot(x, 1 / x, label=r"$\dfrac{1}{x}$")
plt.grid(True)
plt.legend(loc="upper left")
plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x), label=r"$e^x$")
plt.grid(True)
plt.legend(loc="upper left")
plt.show()
```

### gca() and moving the axis

```python
x = np.linspace(-np.pi, np.pi)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
# set the origin with get current axis
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
plt.grid()  # add grid to make it nicer
plt.show()
```
